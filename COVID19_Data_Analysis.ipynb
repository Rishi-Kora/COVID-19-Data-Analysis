Standardizing the values which have "int64" and "float64" data type

````python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
````

When working with machine learning, scaling is an important step because different features (columns) in a dataset might have different ranges. **If one feature has values from 0 to 1000 while another has values from 0 to 1, models might give more importance to the larger numbers.** Scalers help normalize these differences.

There are two ways to Normalize the data
1. StandardScaler (Z-score Normalization)
It transforms the data so that it has a mean of 0 and a standard deviation of 1.
2. MinMaxScaler (Min-Max Normalization)
It scales values between 0 and 1 (or a custom range like -1 to 1).


````python
scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
````

Top 10 affected countries by confirmed cases (latest date)

````python
latest_date = df['Date'].max()
df_latest = df[df['Date'] == latest_date]
top_countries = df_latest.groupby('Country/Region')[['Confirmed']].sum().sort_values(by='Confirmed', ascending=False).head(10)
st.write("### Top 10 Countries by Confirmed Cases:")
st.write(top_countries)
````


Plot top 10 affected countries

Writing the dataset in streamlit to create a dashboard

````python
%%writefile app.py
import pandas as pd
import plotly.express as px
import streamlit as st

# Load dataset
file_path = "/content/covid_19_clean_complete.csv"
df = pd.read_csv(file_path)

# Display basic information
st.title("COVID-19 Data Analysis")
st.write("### Dataset Info:")
st.write(df.info())
st.write("### Summary Statistics:")
st.write(df.describe())

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Aggregate data by date
agg_df = df.groupby('Date')[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()

# Plot the trend of confirmed, deaths, and recovered cases
df_melted = agg_df.melt(id_vars=['Date'], value_vars=['Confirmed', 'Deaths', 'Recovered'],
                         var_name='Case Type', value_name='Count')
fig = px.line(df_melted, x='Date', y='Count', color='Case Type', title="Global COVID-19 Trend Over Time")
st.plotly_chart(fig)

# Top 10 affected countries by confirmed cases (latest date)
latest_date = df['Date'].max()
df_latest = df[df['Date'] == latest_date]
top_countries = df_latest.groupby('Country/Region')[['Confirmed']].sum().sort_values(by='Confirmed', ascending=False).head(10)
st.write("### Top 10 Countries by Confirmed Cases:")
st.write(top_countries)

# Plot top 10 affected countries
fig_bar = px.bar(top_countries, x=top_countries.index, y='Confirmed', title="Top 10 Countries by Confirmed COVID-19 Cases", color='Confirmed', color_continuous_scale='Reds')
st.plotly_chart(fig_bar)
````
